```python
def getDecoys_neg(df, d1_max, d1_min, d2_max, d2_min):
    
    # Get range difference between max-min attributes
    r1 = d1_max - d1_min 
    r2 = d2_max - d2_min 
    
    # Get item 'A'
    A1 = d1_min + (r1/3) 
    A2 = d2_max - (r2/3)
    A = "(" + str(round(A1, 2)) + ", " + str(round(A2, 2)) + ")"
    
    # Get item 'B'
    B1 = d1_max - (r1/3)
    B2 = d2_min + (r2/3)
    B = "(" + str(round(B1, 2)) + ", " + str(round(B2, 2)) + ")"
    
    # Get Phantom decoy A (P_A)
    P_A1 = d1_min + (r1/5)
    P_A2 = d2_max - (r2/3)
    P_A = "(" + str(round(P_A1, 2)) + ", " + str(round(P_A2, 2)) + ")"
    
    # Get P_B
    P_B1 = d1_max - (r1/3)
    P_B2 = d2_min + (r2/5)
    P_B = "(" + str(round(P_B1, 2)) + ", " + str(round(P_B2, 2)) + ")"
    
    # Get Attraction decoy A (A_A)
    A_A1 = A1 + (r1/6)
    A_A2 = A2
    A_A = "(" + str(round(A_A1, 2)) + ", " + str(round(A_A2, 2)) + ")"
    
    # Get A_B
    A_B1 = B1
    A_B2 = B2 + (r2/6)
    A_B = "(" + str(round(A_B1, 2)) + ", " + str(round(A_B2, 2)) + ")"

    # Get Similarity decoy A (S_A)
    S_A1 = A1 + (r1/15)
    S_A2 = A2 - (r2/15)
    S_A = "(" + str(round(S_A1, 2)) + ", " + str(round(S_A2, 2)) + ")"
    
    # Get S_B 
    S_B1 = B1 - (r1/15)
    S_B2 = B2 + (r2/15)
    S_B = "(" + str(round(S_B1, 2)) + ", " + str(round(S_B2, 2)) + ")"
    
    # Get Compromise decoy A (C_A)
    C_A1 = d1_min
    C_A2 = d2_max
    C_A = "(" + str(round(C_A1, 2)) + ", " + str(round(C_A2, 2)) + ")"
    
    # Get C_B 
    C_B1 = d1_max
    C_B2 = d2_min
    C_B = "(" + str(round(C_B1, 2)) + ", " + str(round(C_B2, 2)) + ")"
    
    decoy_data = pd.DataFrame(data={'A': A, 'B': B, 'P_A': P_A, 'P_B': P_B, 'A_A': A_A, 'A_B': A_B, 'S_A': S_A, 'S_B': S_B, 'C_A': C_A, 'C_B': C_B}, index=range(1))
    df = df.append(decoy_data, ignore_index = True)
    
    return df

def getDecoys_pos(df, d1_max, d1_min, d2_max, d2_min):
    
    # Get range difference between max-min attributes
    r1 = d1_max - d1_min 
    r2 = d2_max - d2_min 
    
    #c = (1/3) 
    
    # Get item 'A'
    A1 = d1_min + (r1/3) 
    A2 = d2_max - (r2/3)
    A = "(" + str(round(A1, 2)) + ", " + str(round(A2, 2)) + ")"
    
    # Get item 'B'
    B1 = d1_max - (r1/3)
    B2 = d2_min + (r2/3)
    B = "(" + str(round(B1, 2)) + ", " + str(round(B2, 2)) + ")"
    
    # Get Phantom decoy A (P_A)
    P_A1 = d1_min + (r1/3)
    P_A2 = d2_max - (r2/5)
    P_A = "(" + str(round(P_A1, 2)) + ", " + str(round(P_A2, 2)) + ")"
    
    # Get P_B
    P_B1 = d1_max - (r1/5)
    P_B2 = d2_min + (r2/3)
    P_B = "(" + str(round(P_B1, 2)) + ", " + str(round(P_B2, 2)) + ")"
    
    # Get Attraction decoy A (A_A)
    A_A1 = A1 - (r1/6)
    A_A2 = A2
    A_A = "(" + str(round(A_A1, 2)) + ", " + str(round(A_A2, 2)) + ")"
    
    # Get A_B
    A_B1 = B1
    A_B2 = B2 - (r2/6)
    A_B = "(" + str(round(A_B1, 2)) + ", " + str(round(A_B2, 2)) + ")"

    # Get Similarity decoy A (S_A)
    S_A1 = A1 + (r1/15)
    S_A2 = A2 - (r2/15)
    S_A = "(" + str(round(S_A1, 2)) + ", " + str(round(S_A2, 2)) + ")"
    
    # Get S_B 
    S_B1 = B1 - (r1/15)
    S_B2 = B2 + (r2/15)
    S_B = "(" + str(round(S_B1, 2)) + ", " + str(round(S_B2, 2)) + ")"
    
    # Get Compromise decoy A (C_A)
    C_A1 = d1_min
    C_A2 = d2_max
    C_A = "(" + str(round(C_A1, 2)) + ", " + str(round(C_A2, 2)) + ")"
    
    # Get C_B 
    C_B1 = d1_max
    C_B2 = d2_min
    C_B = "(" + str(round(C_B1, 2)) + ", " + str(round(C_B2, 2)) + ")"
    
    decoy_data = pd.DataFrame(data={'A': A, 'B': B, 'P_A': P_A, 'P_B': P_B, 'A_A': A_A, 'A_B': A_B, 'S_A': S_A, 'S_B': S_B, 'C_A': C_A, 'C_B': C_B}, index=range(1))
    df = df.append(decoy_data, ignore_index = True)
    
    return df
    
```
